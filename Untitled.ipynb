# Pertama kita import dulu library yang akan digunakan pandas, matplotlib, seaborn, dan numpy

```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib
import seaborn as sns
import numpy as np

matplotlib.style.use('ggplot')
```

# kemudian kita read datanya, disini saya menggunakan decoder ISO-8859-1 karena ada error jika menggunakan decoder default (UTF-8)

# kemudian untuk mencari data yang ingin divisualisasikan saya menggunakan df.head() untuk melihat kolom yang sekiranya bisa divisualkan

```python
dataFrame = pd.read_csv('csgo_round_snapshots.csv')
dataFrame.head()
```

# df.shape untuk melihat banyaknya row dan col yang ada

```python
dataFrame.shape
```

```python
dataFrame.set_index('time_left', inplace=True) 
```

# BAR PLOT

# visualisasi data menggunakan bar plot

```python
jumlah_ct_score = dataFrame['ct_score'].value_counts()
jumlah_ct_score.plot.bar()
plt.title('Perbandingan Jumlah sisa waktu per round')
```

```python
jumlah_ct_score.plot(kind='bar')
```

```python
ct_score_group = dataFrame.groupby('ct_score').sum()
ct_score_group.head()
```

```python
ct_score_group['ct_money'].plot(kind='bar')
```

```python
plt.bar(ct_score_group.index,
        ct_score_group['ct_money'])
plt.xticks(rotation=60, ha='right')
```

```python
sns.barplot(x=ct_score_group.index,
            y=ct_score_group['ct_money'])
plt.xticks(rotation=60, ha='right')
```

# setiap round dan score dari tim counter terorist yang mana uang terbanyak di dapatkan di score 6.0

# Histogram

```python
dataFrame['ct_money'].plot.hist(range=(0, 1000))
```

```python
sns.histplot(dataFrame['ct_money'], bins=10, kde=False, binrange=(0, 1000))
```

# ct_money /  pendatapan per round dari counter terorist yang paling rendah ada di 1200$

# PIE CHART

```python
ct_score_group.index
```

```python
plt.figure(figsize=(50, 10))
plt.pie(ct_score_group['ct_grenade_flashbang'].sort_values(),
        labels=ct_score_group['ct_grenade_flashbang'].sort_values().index,
        autopct='%.2f%%',
        startangle=90,)
plt.show()
```

# presentase penggunaan flashbang tim counter terorisr terbanyak di score 3.0

# Visualisasi Menggunakan HEATMAP

```python
round_total_corr = dataFrame.loc[:, 'ct_score':'t_score'].corr()
round_total_corr
```

```python
round_total_corr.style.background_gradient(cmap='coolwarm').set_properties(**{'font-size': '20px'})
```

```python
plt.imshow(round_total_corr, cmap='coolwarm')
plt.colorbar()
plt.xticks(range(len(round_total_corr)), round_total_corr.columns)
plt.yticks(range(len(round_total_corr)), round_total_corr.columns)
plt.grid(False)
```

```python
dataFrame.head()
```

```python
sns.heatmap(
    data=round_total_corr,
    annot=True
)
```

# Pertarungan tim counter terorist dengan teroris sangat sengit
